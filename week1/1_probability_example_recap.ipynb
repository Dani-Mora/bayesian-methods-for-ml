## Thief-Alarm example

We haven an alarm (A) system intended to be triggered when a thief (T) breaks in. Alarm may also trigger in case of an earthquake (E). When an eartquake occurs, there might be a radio announcement (R).

Let's plot the graph:

```python
ex = Digraph()
ex.node('A')
ex.node('T')
ex.node('R')
ex.node('E')
ex.edges(['TA', 'EA', 'ER'])
ex
```

Its joint probability is defined as:
    
$$
P(T, A, E, R) = P(A|E, T) P(R|E) P(T) P(E)
$$

We want to know which is the probability of a thief breaking in (i.e. T=1) given the alarm has been triggered (i.e. A=1)? We want to compute P(T=1|A=1). For simplicity, we will use $T$ for $T=1$ and $\bar{T}$ for $T=0$.

Using conditional probability definition:

$$
P(T|A) = \frac{P(T,A)}{P(A)}
$$

Then, using sum rule, we have:

$$
\begin{align*}
P(T|A) & =  \frac{\sum_{e\in \{0, 1\}} P(T, A, E=e)}{\sum_{e\in \{0, 1\}, t \in \{0, 1\}} P(T=t, A, E=e)} \\
P(T|A) & =  \frac{P(T,A,E) + P(T,A,\bar{E})}{P(\bar{T}, A, \bar{E}) + P(T, A, \bar{E}) + P(\bar{T}, A, E) + P(T, A, E)}
\end{align*}
$$

Where:

$$
\begin{align*}
P(T,A,E) & = P(A|E,T)P(T)P(E) \\
P(T,A,\bar{E}) & = P(A|\bar{E},T)P(T)P(\bar{E}) \\
\ldots \\
\end{align*}
$$

Computing $P(\bar{T}, A, \bar{E})$ and $P(\bar{T}, A, E)$ is straightforward.

What if we now listen a radio announcement that there has been an earthquake in the area?

We want to compute now $P(T|A,R)$.

Using conditional probability definition:

$$
P(T|A,R) = \frac{P(T,A,R)}{P(A,R)}
$$

Using sum rule, we have:

$$
\begin{align*}
P(T|A,R) & = \frac{\sum_{e\in \{0, 1\}} P(T, A, E=e, R)}{\sum_{e\in \{0, 1\}, t\in \{0, 1\}} P(T=t, A, E=e, R)} \\
P(T|A,R) & = \frac{P(T, A, E, R) + P(T, A, \bar{E}, T)}{P(\bar{T}, A, \bar{E}, R) + P(T, A, \bar{E}, R) + P(\bar{T}, A, E, R) + P(T, A, E, R)} \\
\end{align*}
$$

Where we can compute, from the joint probability distribution:

$$
P(T, A, E, R) = P(A|E, T) P(R|E) P(T) P(E)
$$

The rest of the terms are straightforward to compute.
